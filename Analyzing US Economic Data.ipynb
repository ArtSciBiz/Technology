```python
import types
import pandas as pd

# Point to the CSV file that resides in an cloud object storage(COR) bucket that is made publicly accessible on the Web (data is public)
csvFile_unemploymentData = 'https://datasciencepythonbasicsproject-donotdelete-pr-raugcxqfl1o6nv.s3.us.cloud-object-storage.appdomain.cloud/BureauOfLaborStatistics_UnemploymentData.csv'

# Read CSV file using pandas read_csv method, & retrieve unemployment data from the file 
dataFrame = pd.read_csv(csvFile_unemploymentData)

# Use DataFrame.query() to select UNEMPLOYMENT_PERCENT, column values that are greater than 8.5
selectedData = dataFrame.query('UNEMPLOYMENT_PERCENT > 8.5')

# Considering the current low unemployment rate, go back to year when the unemployment was above 8.5% (sort descending year) 
selectedData.sort_values(['YEAR', 'MONTH'], ascending=[0,1])
```

```python
import types
import pandas as pd

# Create a custom function called make_dashboard that takes 4 parameters (MBehara 2019/12/09)
def make_dashboard(gdp_year, gdp_value, unemployment_year, unemployment_value):
    from matplotlib import pyplot as plt

    fig, (ax1, ax2) = plt.subplots(2)   #create 2 subplots

    fig.suptitle('Data Souce: US Bureau of Economic Analysis & Labor Statistics \n(MB20191209)')
    
    ax1.set_xlabel('YEARS')
    ax1.set_ylabel('GDP US$')
    ax1.plot(gdp_year, gdp_value, 'tab:green')

    ax2.set_xlabel('YEARS')
    ax2.set_ylabel('UNEMPLOYMENT %')
    ax2.plot(unemployment_year, unemployment_value, 'tab:red')

# Point to the CSV file that resides in the IBM cloud object storage(COR) bucket that is made publicly accessible on the Web (data is public)
csvFile_gdpData = 'https://datasciencepythonbasicsproject-donotdelete-pr-raugcxqfl1o6nv.s3.us.cloud-object-storage.appdomain.cloud/BureauOfEconomicAnalysis_GDP_Annual.csv'

# Using pandas library, read CSV file into a dataframe using read_csv method, & retrieve the retrieve gross domestic product data from the file 
df_gdpData = pd.read_csv(csvFile_gdpData)

gdp_yr = df_gdpData[['YEAR']]
gdp = df_gdpData[['GDP_BILLIONS']]

# Point to the CSV file that resides in the IBM cloud object storage(COR) bucket that is made publicly accessible on the Web (data is public)
csvFile_unemploymentData = 'https://datasciencepythonbasicsproject-donotdelete-pr-raugcxqfl1o6nv.s3.us.cloud-object-storage.appdomain.cloud/BureauOfLaborStatistics_UNEMPLOYMENT_AnnualMax.csv'

# Using pandas library, read CSV file into a dataframe using read_csv method, & retrieve the unemployment data from the file 
df_unemploymentData = pd.read_csv(csvFile_unemploymentData)

unemployment_yr = df_unemploymentData[['YEAR']]
unemployment = df_unemploymentData[['UNEMPLOYMENT_PERCENT']]


make_dashboard(gdp_yr, gdp, unemployment_yr, unemployment)
```
